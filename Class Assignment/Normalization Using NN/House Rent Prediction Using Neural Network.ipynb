```python
#Libraries
from numpy import loadtxt
from sklearn.model_selection import train_test_split
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU,PReLU,ELU
from keras.layers import Dropout
import pandas as pd
import matplotlib.pyplot as plt
import csv
import numpy as np
import seaborn as sns
from google.colab import drive
drive.mount('/content/drive')
```

```python
#Dataset loading and preprocessing
data_path = '/content/drive/MyDrive/Colab Notebooks/nn-data/House_Rent_Dataset.csv'
df = pd.read_csv(data_path)
df.shape
```

```python
# Convert the entire DataFrame to numeric
# X = X.apply(pd.to_numeric, errors='coerce')#alternative

df_numeric = df._get_numeric_data()
df_numeric
```

```python
# split into input (X) and output (y) variables
# split into input (X) and output (y) variables
# Selecting features and target
X = df_numeric.iloc[:, [0, 2, 3]]  # Assuming 'BHK', 'Size', and 'Bathroom' are the first, third, and fourth columns
y = df_numeric.iloc[:,1]  # Assuming 'Rent' is the second column
```

```python
y.head()
```

```python
X.head()
```

```python
# Normalize features using Min-Max Scaling
from sklearn.preprocessing import MinMaxScaler,StandardScaler
# 1st way
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# 2nd way
# scaler = StandardScaler()
# X_scaled = scaler.fit_transform(X)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
y_train.head()
```

```python
X_test.shape
```

```python
y_test.head()
```

```python
# Build the neural network model
model = Sequential()
model.add(Dense(16, input_dim=3, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='linear'))  # For regression, linear activation is used
```

```python
# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.2, verbose=1)
```

```python
# evaluate the keras model
_, accuracy = model.evaluate(X_train, y_train,)
print('Accuracy: %.2f' % (accuracy*100))
```

```python
predictions=model.predict(X_test)
```
